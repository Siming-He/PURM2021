```python
from queue import PriorityQueue
import numpy as np
import matplotlib.pyplot as plt
from heapq import heapify, heappush, heappop
```

```python
map = np.random.randint(-5, 30, size = (10, 10))
```

```python
class Coordinate:
    def __init__(self, value, x, y):
        self.value = value
        self.x = x
        self.y = y
        self.piX = -1
        self.piY = -1
        self.visited = False
        
    def setVal(self, value):
        self.value = value
        
    def setPi(self, x, y):
        self.piX = x
        self.piY = y
```

```python
class Dijkstra:
    def __init__(self, map, srcX, srcY):
        self.index = 0
        self.map = map
        self.srcX = srcX
        self.srcY = srcY
        self.dstMap = np.full((map.shape[0], map.shape[1]), None)
        for i in range(0, map.shape[0]):
            for j in range(0, map.shape[1]):
                self.dstMap[j][i] = Coordinate(np.inf, i, j)
        self.heap = []
        heapify(self.heap)
        self.dijkstra()
    
    def getShortestPath(self, tgtX, tgtY):
        tgt = self.dstMap[tgtY][tgtX]
        
        path = []
        path.insert(0, (tgt.x, tgt.y))
        
        while (tgt.piX != -1):
            tgt = self.dstMap[tgt.piY][tgt.piX]
            path.insert(0, (tgt.x, tgt.y))
            if (tgt.x == self.srcX and tgt.y == self.srcY):
                break
        if (path[0] == (self.srcX, self.srcY)):
            return path
        else:
            return None
            
    def dijkstra(self):
        self.dstMap[self.srcY][self.srcX].setVal(0)
        heappush(self.heap, (self.dstMap[self.srcY][self.srcX].value, self.index, self.dstMap[self.srcY][self.srcX]))
        self.index = self.index + 1
        
        i = 0
        while len(self.heap) > 0:
            u = heappop(self.heap)
            #print(u.x, u.y, u.value)
            u = u[2]
            u.visited = True
            
            if (u.y - 1 >= 0):
                if (self.map[u.y - 1][u.x] > 0 and not self.dstMap[u.y - 1][u.x].visited):
                    self.cmpDist(u, self.dstMap[u.y - 1][u.x])
                    
            if (u.y + 1 < map.shape[0]):
                if (self.map[u.y + 1][u.x] > 0 and not self.dstMap[u.y + 1][u.x].visited):
                    self.cmpDist(u, self.dstMap[u.y + 1][u.x])
                
            if (u.x - 1 >= 0):
                if (self.map[u.y][u.x - 1] > 0 and not self.dstMap[u.y][u.x - 1].visited):
                    self.cmpDist(u, self.dstMap[u.y][u.x - 1])
                
            if (u.x + 1 < map.shape[1]):
                if (self.map[u.y][u.x + 1] > 0 and not self.dstMap[u.y][u.x + 1].visited):
                    self.cmpDist(u, self.dstMap[u.y][u.x + 1])
                    
            #print(u.x, u.y+1, self.dstMap[u.y+1][u.x].value)
            #print(u.x+1, u.y, self.dstMap[u.y][u.x+1].value)
                    
    def cmpDist(self, u,  v):
        if (v.value > u.value + self.map[v.y][v.x]):
            v.setVal(u.value + self.map[v.y][v.x])
            v.setPi(u.x, u.y)
            heappush(self.heap, (v.value, self.index, v))
            self.index = self.index + 1
            #print(self.heap[0][2].x, self.heap[0][2].y, self.heap[0][2].value)
```

```python
d = Dijkstra(map, 0, 0)
path = d.getShortestPath(8, 8)

newMap = map.copy() + 15

for p in path:
    #print(p)
    newMap[p[1]][p[0]] = -1000


fig, ax = plt.subplots(figsize = (8, 8))
ax.matshow(newMap, cmap = "RdBu", vmin = 0, vmax = 20)

for i in range(10):
    for j in range(10):
      val = map[j,i]#d.dstMap[j][i].value #(d.dstMap[j][i].piX, d.dstMap[j][i].piY) # 
      ax.text(i, j, str(val), va='center', ha='center')
```
